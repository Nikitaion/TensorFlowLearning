```python
# Import the car sales DataFrame with missing data ("car-sales-missing-data.csv")
car_sales_missing = pd.read_csv("/content/car-sales-missing-data.csv")
# Check out the new DataFrame
car_sales_missing
```

```python
# Fill the Odometer column missing values with the mean of the column inplace
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(), inplace=True)
```

```python
# View the car sales missing DataFrame and verify the changes
car_sales_missing
```

```python
# Remove the rest of the missing data inplace

# Before in need change Price to integer and remove NaN
car_sales_missing["Price"] = car_sales_missing["Price"].str.replace("[\$\,\.]", "")
car_sales_missing["Price"] = car_sales_missing['Price'].fillna(0).astype(int)
car_sales_missing
```

```python
# Change zeros to mean value(value=...) (exclude zeros)
car_sales_missing["Price"] = car_sales_missing["Price"].replace(to_replace=0, value=car_sales_missing["Price"].replace(0, np.NaN).mean())
# Check changes
car_sales_missing["Price"]
```

```python
# Remove the rest of the missing data inplace
car_sales_missing["Doors"].fillna(car_sales_missing["Doors"].mean(), inplace=True)
# Check changes
car_sales_missing
```

```python
# We'll now start to add columns to our DataFrame.

# Create a "Seats" column where every row has a value of 5
car_sales_missing['Seats'] = [5] * len(car_sales_missing)
car_sales_missing
```

```python
# Create a column called "Engine Size" with random values between 1.3 and 4.5
# Remember: If you're doing it from a Python list, the list has to be the same length
# as the DataFrame
import random
engine_size = []
i=0
while i < len(car_sales_missing):
  engine_size += [round(random.uniform(1.3, 4.5), 1)]
  i += 1

car_sales_missing['Engine Size'] = engine_size
car_sales_missing
```
